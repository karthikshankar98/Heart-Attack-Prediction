```r
split = sample.split(df$output, SplitRatio = 0.70)
train = subset(df, split == TRUE)
test = subset(df, split == FALSE)
```

## Logistic Regression

```r
log.model <- glm(formula=output ~ . , family = binomial(link='logit'),data = train)

fitted.probabilities <- predict(log.model,newdata=test,type='response')
fitted.results <- ifelse(fitted.probabilities > 0.5,1,0)
misClasificError <- mean(fitted.results != test$output)

print(paste('Accuracy',1-misClasificError))
table(test$output, fitted.probabilities > 0.5)
```

## Random Forest

```r
random.model <- randomForest(output ~ ., data = train)
print(random.model)
```
